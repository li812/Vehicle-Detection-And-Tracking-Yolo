```python
import cv2
from ultralytics import YOLO
model = YOLO('yolov8x.pt') # yolov3-v7
model.names
cap = cv2.VideoCapture('scan_video/120223KLM.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detect vehicles in the frame
    results = model(frame)

    # Draw bounding boxes on the frame
    for result in results:
        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = box
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Vehicle Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


modify this code. Just detect vehicles from the video and don't show the video playback. save the frames with vehicles in a folder called 'detected_vehicles'
```
